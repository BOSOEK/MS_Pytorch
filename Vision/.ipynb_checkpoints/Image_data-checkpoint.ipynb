# **이미지 데이터 처리 소개**

```python
import torch
import torchvision
import matplotlib.pyplot as plt
import numpy as np
from torchvision.transforms import ToTensor
import torch.nn as nn

data_train = torchvision.datasets.MNIST('./data', download=True, train=True, transform=ToTensor())
data_test = torchvision.datasets.MNIST('./data', download=True, train=False, transform=ToTensor())

# 데이터 시각화
fig, ax = plt.subplots(1, 7)
for i in range(7):
    ax[i].imshow(data_train[i][0].view(28, 28))
    ax[i].set_title(data_train[i][1])
    ax[i].axis('off')
```

# **단순 조밀 신경망 훈련**
이미지 분류를 위한 **완전 연결 신경망(퍼셉트론)** 시작

```python
# 신경망 정의
net = nn.Sequential(
        nn.Flatten(),
        nn.Linear(784, 10),
        nn.LogSoftmax())
# 신경망 훈련
print('예측할 숫자 : ', data_train[0][1])
torch.exp(net(data_train[0][0]))
```

```python
# 훈련되지 않아서 각각의 확률이 비슷하다
# 모델 학습을 위해 데이터 세트 배치를 만든다
train_loader = torch.utils.data.DataLoader(data_train, batch_size=64)
test_loader = torch.utils.data.DataLoader(data_test, batch_size=64)
```

**학습 프로세스 단계**
1. 엡력 데이터와 라벨로 구성된 데이터 세트에서 미니 배치 가져옴
2. 미니 배치에 대한 예측 결과 계산
3. 결과와 예상 결과의 차이는 **손실 함수** 라는 특수 함수로 계산
4. 모델 가중치와 관련해 손실함수의 기울기를 계산한 후 가중치로 조정하여 네트워크의 성능을 최적화 하는데 사용. 
5. 전체 데이터 처리까지 반복하며 데이터 세트를 통한 하나의 완전한 패스를 __epochs__ 라고 함.

```python
# 훈련 수행 코드
def train_epoch(net,dataloader,lr=0.01,optimizer=None,loss_fn = nn.NLLLoss()):
    # 매개변수 : 신경망, 학습 데이터, 최적화 알고리즘, 학습률, 예상 결과와 생성 결과의 차이 측정 함수
    optimizer = optimizer or torch.optim.Adam(net.parameters(),lr=lr) # 최적화 알고리즘 아담 사용
    net.train()  # 네트워크를 훈련모드로 전환
    total_loss,acc,count = 0,0,0
    for features,labels in dataloader:
        optimizer.zero_grad()
        out = net(features)
        loss = loss_fn(out,labels) #cross_entropy(out,labels)
        loss.backward()
        optimizer.step()
        total_loss+=loss
        _,predicted = torch.max(out,1)
        acc+=(predicted==labels).sum()
        count+=len(labels)
    return total_loss.item()/count, acc.item()/count

train_epoch(net,train_loader)
```

**훈련 작업 순서**
* 네트워크를 훈련 모드로 전환(`net.train()`)
* 데이터 세트의 모든 배치를 검토하고 각 배치에 대해 다음을 수행
    * 배치(`out`)에 대한 네트워크의 예측 계산
    * 예측과 기대값의 불일치(`loss`) 계산
    * 네트워크 가중치 조정으로 손실 최소화(`optimizer.step()`)
    * 정확하게 예측된 케이스 수를 계산(**정확도**)

```python
# 네트워크 평가 코드
def validate(net, dataloader,loss_fn=nn.NLLLoss()):
    net.eval()
    count,acc,loss = 0,0,0
    with torch.no_grad():
        for features,labels in dataloader:
            out = net(features)
            loss += loss_fn(out,labels) 
            pred = torch.max(out,1)[1]
            acc += (pred==labels).sum()
            count += len(labels)
    return loss.item()/count, acc.item()/count

validate(net,test_loader)
```

```python
# 훈련과 검증 모두에 사용 가능한 코드
def train(net,train_loader,test_loader,optimizer=None,lr=0.01,epochs=10,loss_fn=nn.NLLLoss()):
    optimizer = optimizer or torch.optim.Adam(net.parameters(),lr=lr)
    res = { 'train_loss' : [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
    for ep in range(epochs):
        tl,ta = train_epoch(net,train_loader,optimizer=optimizer,lr=lr,loss_fn=loss_fn)
        vl,va = validate(net,test_loader,loss_fn=loss_fn)
        print(f"Epoch {ep:2}, Train acc={ta:.3f}, Val acc={va:.3f}, Train loss={tl:.3f}, Val loss={vl:.3f}")
        res['train_loss'].append(tl)
        res['train_acc'].append(ta)
        res['val_loss'].append(vl)
        res['val_acc'].append(va)
    return res

# 네트워크 초기화 후 다시 시작
net = nn.Sequential(
        nn.Flatten(), 
        nn.Linear(784,10), # 784 inputs, 10 outputs
        nn.LogSoftmax())

hist = train(net,train_loader,test_loader,epochs=5)
```

```python
# 자 그래프로 봐보장
def plot_results(hist):
    plt.figure(figsize=(15,5))
    plt.subplot(121)
    plt.plot(hist['train_acc'], label='Training acc')
    plt.plot(hist['val_acc'], label='Validation acc')
    plt.legend()
    plt.subplot(122)
    plt.plot(hist['train_loss'], label='Training loss')
    plt.plot(hist['val_loss'], label='Validation loss')
    plt.legend()
plot_results(hist)
```

> 그래프로 봐보면 과적합 됐다는 것을 알 수 있다
### **네트워크 가중치 시각화**
네트워크가 **하나의 레이어보다 더 복잡** 하면 시각화 하기 어렵지만 위 경우는 하나라 시각화가 가능하다.   
`weight_tensor`에 `net.parameters()`로 신경망의 입력 크기의 텐서를 만들고 숫자가 0인지 아닌지 확인하려면 입력 숫자에 `weight_tensor[0]`을 곱하고 결과를 소프트 맥스 함수로 얻는다. 결과가 가중치 텐서 요소가 분류하는 숫자의 평균 모양과 다소 유사합니다.


```python
weight_tensor = next(net.parameters())
fig,ax = plt.subplots(1,10,figsize=(15,4))
for i,x in enumerate(weight_tensor):
    ax[i].imshow(x.view(28,28).detach())
```
